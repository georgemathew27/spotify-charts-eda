```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("albums.csv")
print(df.shape)
print(df.head())
```

```python
df.info()
df.isnull().sum()
```

```python
top_artists = df['artist_names'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.values, y=top_artists.index)
plt.xlabel("Number of Albums")
plt.ylabel("Artist")
plt.title("Top 10 Artists by Album Count")
plt.show()
```

```python
df['release_date']=pd.to_datetime(df['release_date'], errors='coerce')
df['release_year']=df['release_date'].dt.year
albums_per_year = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 5))
sns.lineplot(x=albums_per_year.index, y=albums_per_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Albums")
plt.title("Albums Released per Year")
plt.show()
```

```python
top_labels = df['label'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_labels.values, y=top_labels.index)
plt.xlabel("Number of Albums")
plt.ylabel("Label")
plt.title("Top 10 Labels by Album Count")
plt.show()
```
